```python
import pandas as pd

df = pd.read_feather('exp2_glue_scores.arrow') * 100
```

```python
GLUE_SCORE_FULL = [
    ('mrpc', 'eval_combined_score'),
    ('cola', 'eval_matthews_correlation'),
    ('mnli', 'eval_accuracy'),
    ('mnli', 'eval_accuracy_mm'),
    ('qqp', 'eval_combined_score'),
    ('rte', 'eval_accuracy'),
    ('stsb', 'eval_combined_score'),
    ('qnli', 'eval_accuracy'),
    ('sst2', 'eval_accuracy'),
    ('wnli', 'eval_accuracy'),
]
GLUE_SCORE_BERT = [
    ('mrpc', 'eval_combined_score'),
    ('cola', 'eval_matthews_correlation'),
    ('mnli', 'eval_accuracy'),
    ('mnli', 'eval_accuracy_mm'),
    ('qqp', 'eval_combined_score'),
    ('rte', 'eval_accuracy'),
    ('stsb', 'eval_combined_score'),
    ('qnli', 'eval_accuracy'),
    ('sst2', 'eval_accuracy'),
]

df[('glue_score', 'full')] = df[GLUE_SCORE_FULL].apply(lambda x: x.mean(), axis=1)
df[('glue_score', 'bert')] = df[GLUE_SCORE_BERT].apply(lambda x: x.mean(), axis=1)
```
